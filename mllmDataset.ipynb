```python
def save_to_csv(article, output1, output2, output3, path):
    """
    Appends a row of data (Test article, Prompt1, Prompt2, Prompt3) to a CSV file.
    The function automatically checks if the file exists to write the header only once.

    Args:
        article (str): The text of the test article.
        output1 (str): The output from Prompt1.
        output2 (str): The output from Prompt2.
        output3 (str): The output from Prompt3.
        path (str): The full path to the CSV file.
    """
    os.makedirs(os.path.dirname(path), exist_ok=True)  # Ensure the directory exists

    # Check if the file exists BEFORE opening in append mode
    # This determines if the header needs to be written.
    file_already_exists = os.path.exists(path) and os.path.getsize(path) > 0

    with open(path, "a", encoding="utf-8", newline="") as file:
        writer = csv.writer(file, quoting=csv.QUOTE_ALL)
        writer.writerow([article, output1, output2, output3])
```
